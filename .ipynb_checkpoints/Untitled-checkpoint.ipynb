```python
# transform target variable to encoding
from sklearn.preprocessing import LabelEncoder

# creating an instance of label encoder
labelencoder=LabelEncoder()

# assign numerical values and storing it in another column called class_encoder.
beans_df['class_encoder']=labelencoder.fit_transform(beans_df['y'])

beans_df.head()
```

```python
# this to check the frequncy of Class and Class encoder and their frequencies.
import sidetable
beans_df.stb.freq(['y','class_encoder'])
```

```python
# drop the original target variable.
beans_df.drop(['y'],inplace=True, axis=1)
```

```python
# view the data.
beans_df.head()
```

```python
# apply Robust scaler
from sklearn.preprocessing import RobustScaler
trans = RobustScaler()
rb_df = trans.fit_transform(beans_df)
# convert the array back to a dataframe
scaled_df = pd.DataFrame(rb_df,columns=beans_df.columns)
scaled_df.head()
```

```python
pca1 = PCA(n_components= 7, random_state= 22)
pca_1 = pca1.fit_transform(scaled_df)
```

```python
pca_df = pd.DataFrame(pca_1, columns=['f1', 'f2', 'f3', 'f4', 'f5', 'f6', 'f7'])
pca_df['class_encoder'] = beans_df['class_encoder']
pca_df
```

```python
df_train, df_val = train_test_split(pca_df, test_size=0.2, random_state=42)

X_train = df_train.drop(columns=['class_encoder'])
y_train = df_train[['class_encoder']]

X_val = df_val.drop(columns=['class_encoder'])
y_val = df_val[['class_encoder']]

print(f'X_train shape is: {X_train.shape}.')
print(f'X_val shape is: {X_val.shape}.')
print(f'y_train shape is: {y_train.shape}.')
print(f'y_val shape is: {y_val.shape}.')
```

```python
classifier1 = KNeighborsClassifier()

classifier1 = classifier1.fit(X_train, y_train)
y_pred = classifier1.predict(X_val)
```

```python
#Estimating accuracy using accuracy score

acc1 = accuracy_score(y_val, y_pred)

print (f'Accuracy = {acc1}')
```

```python
classifier4 = SVC(random_state=0, probability=True)

classifier4 = classifier4.fit(X_train, y_train)
y_pred = classifier4.predict(X_val)
```

```python
#Estimating accuracy using accuracy score

acc4 = accuracy_score(y_val, y_pred)

print (f'Accuracy = {acc4}')
```
